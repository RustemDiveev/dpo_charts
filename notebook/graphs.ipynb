## Пример 1

Суть в следующем, у меня есть следующие данные:

2024 - 400 образовательных программ
2023 - 300 образовательных программ
2022 - 200 образовательных программ

2024 - 40 сетевых образовательных программ
2023 - 30 сетевых образовательных программ
2022 - 20 сетевых образовательных программ

Мне нужно нарисовать график, показывающий динамику изменения количество образовательных программ разных видов, а также процентное соотношение сетевых образовательных программ от общего числа.

```python
import matplotlib.pyplot as plt
import numpy as np

# Данные
years = [2022, 2023, 2024]
total_programs = [200, 300, 400]
network_programs = [20, 30, 40]

# Рассчет процента сетевых программ
percent_network = [n/t*100 for n, t in zip(network_programs, total_programs)]

# Создание фигуры и основных осей
fig, ax1 = plt.subplots(figsize=(10, 6))

# Настройка позиций для столбцов
x = np.arange(len(years))
width = 0.35

# Столбцы для абсолютных значений
bars1 = ax1.bar(x - width/2, total_programs, width, label='Все программы', color='lightblue')
bars2 = ax1.bar(x + width/2, network_programs, width, label='Сетевые программы', color='steelblue')

# Настройка основной оси Y
ax1.set_xlabel('Годы')
ax1.set_ylabel('Количество программ')
ax1.set_xticks(x)
ax1.set_xticklabels(years)
ax1.legend(loc='upper left')

# Создание второй оси Y для процентов
ax2 = ax1.twinx()
ax2.plot(x, percent_network, color='crimson', marker='o', linewidth=2, label='Доля сетевых программ')
ax2.set_ylabel('Процент сетевых программ (%)')
ax2.set_ylim(0, 15)
ax2.legend(loc='upper right')

# Добавление подписей значений
for i, (tot, net, perc) in enumerate(zip(total_programs, network_programs, percent_network)):
    ax1.text(i - width/2, tot + 10, str(tot), ha='center')
    ax1.text(i + width/2, net + 10, str(net), ha='center')
    ax2.text(i, perc + 0.5, f'{perc:.1f}%', ha='center', color='crimson')

# Дополнительные настройки
plt.title('Динамика образовательных программ и доля сетевых программ')
ax1.grid(axis='y', alpha=0.3)
fig.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Данные
years = [2020, 2021, 2022, 2023, 2024]
total_programs = [211523, 217917, 261058, 272568, 309398]
network_programs = [4885, 6083, 4704, 5212, 5760]

# Рассчет процента сетевых программ
percent_network = [n/t*100 for n, t in zip(network_programs, total_programs)]

# Создание фигуры с 4 подграфиками
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Список заголовков для каждого графика
titles = [
    'Программы повышения квалификации - Регион 1',
    'Программы повышения квалификации - Регион 2', 
    'Программы повышения квалификации - Регион 3',
    'Программы повышения квалификации - Регион 4'
]

# Функция для создания одного графика
def create_plot(ax, title):
    # Настройка позиций для столбцов
    x = np.arange(len(years))
    width = 0.35

    # Столбцы для абсолютных значений
    bars1 = ax.bar(x - width/2, total_programs, width, label='Все программы', color='lightblue')
    bars2 = ax.bar(x + width/2, network_programs, width, label='Сетевая форма', color='steelblue')

    # Настройка основной оси Y
    ax.set_xlabel('Годы')
    ax.set_ylabel('Количество программ')
    ax.set_xticks(x)
    ax.set_xticklabels(years)
    ax.legend(loc='upper left')

    # Создание второй оси Y для процентов
    ax2 = ax.twinx()
    ax2.plot(x, percent_network, color='crimson', marker='o', linewidth=2, label='Доля сетевых программ')
    ax2.set_ylabel('Процент сетевых программ (%)')
    ax2.set_ylim(0, 15)
    ax2.legend(loc='upper right')

    # Добавление подписей значений
    for i, (tot, net, perc) in enumerate(zip(total_programs, network_programs, percent_network)):
        ax.text(i - width/2, tot + 10, str(tot), ha='center', fontsize=8)
        ax.text(i + width/2, net + 10, str(net), ha='center', fontsize=8)
        ax2.text(i, perc + 0.5, f'{perc:.1f}%', ha='center', color='crimson', fontsize=8)

    # Дополнительные настройки
    ax.set_title(title)
    ax.grid(axis='y', alpha=0.3)

# Создаем графики для всех осей
for i in range(2):
    for j in range(2):
        create_plot(axes[i, j], titles[i*2 + j])

# Настраиваем расстояние между подграфиками
plt.tight_layout()

# Показываем график
plt.show()
```
